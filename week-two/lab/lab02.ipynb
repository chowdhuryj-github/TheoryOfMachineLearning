# <span style="color: green;"> Modelling & Model Error </span>
This Jupyter Notebook is created from the course CSC4601 (Theory of Machine Learning) to achieve the following learning outcomes:

- Build confidence and familiarity with model parameters, model input/output, model error and the differences between all of them.
- Interpret and code models from their mathematical notation
- Characterize a model's performance when compared to a given dataset

### <span style="color: green;"> The Aim of the Lab </span>
In this lab, we investigate two commonly used models and the data that these models attempt to replicate. To build familiarity with different parts of the model, we attempt to achieve the following:

- [ ] Identify and visualize the given data, including the data's independent and dependent variables
- [ ] Identify the model you will use to simulate, or predict, that data that you are given
- [ ] Code your model such that it accepts inputs of independent variables and tunable parameters so that it produces a prediction or dependent variable as output
- [ ] Visualize your model's output and plot the model's output against the given data
- [ ] Calculate and summarize the difference, or error, between your model's output and the given data


# <span style="color: green;"> Experiment 1: Gaussian Distribution </span>
In this experiment, we walk through loading and plotting a Gaussian dataset, writing out the Gaussian PDF in a markdown cell, implement the Gaussian model in code, and generating model predictions of several parameter sets to compare against the data. Lastly, we calculate and report the Mean Absolute Error (MAE) for each parameter set and include the errors in a combined plot with a legend.
