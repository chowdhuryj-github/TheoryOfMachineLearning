## Part I: Decision Boundaries and Evaluation of Model Predictions with Metrics
In this part, we load an visualize the  `setosa_data.csv` dataset, plot its points colored by class, and draw the decision boundaries using their equations. We then rewrite each boundary in standard form, classify points using `linear_decision_boundary_classifier()` and evaluate the predictions with `sklearn.metrics.accuracy_score()`.

```python

"""

    a. load the "setosa_data.csv" using the Pandas read_csv() function
    b. display the first few rows of the DataFrame using the head() function

"""

# importing the required libraries
import pandas as pd

# loading the .csv file and displaying the first few rows
setosa_df = pd.read_csv("C:/GitHub/Semester 5/TheoryOfMachineLearning/4_week-four/lab/data/setosa_data.csv")
print(setosa_df.head(10))
```

```python

"""

    c. create a scatter plot of the dataset points with the sepal length on the horizontal axis and sepal width on the vertical axis
        color each point by its class (orange -> setosa & blue -> not setosa)

"""

# importing the required libraries
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# mapping each class label to a color
colors = setosa_df['label'].map(lambda x: 'orange' if x == 'setosa' else 'blue')

# creating a scatter plot
plt.scatter(setosa_df['sepal_length (cm)'], setosa_df['sepal_width (cm)'], c=colors)

# setting up the legend
legend_elements = [Patch(facecolor='orange', label='Setosa'),
                   Patch(facecolor='blue', label='Not Setosa')]
plt.legend(handles=legend_elements)


plt.title("Sepal Length vs Sepal Width for Setosa & Not Setosa")
plt.xlabel("sepal_length (cm)")
plt.ylabel("sepal_width (cm)")
plt.show()

```

```python

"""
    
    d. plot the specified color and label the decision boundaries defined by the following equations:
        i. decision boundary 1 (black): x2 = 2 x1 - 8
        ii. decision boundary 2 (red): x2 = 0.75 x1 - 0.9
        iii. decision boundary 3 (purple): x2 = -0.2 x1 + 12

"""

# importing required libraries
import numpy as np

# i values
x1i = np.linspace(setosa_df['sepal_length (cm)'].min(), setosa_df['sepal_length (cm)'].max(), 100)
x2i = 2 * x1i - 8

# ii values
x1ii = np.linspace(setosa_df['sepal_length (cm)'].min(), setosa_df['sepal_length (cm)'].max(), 100)
x2ii = 0.75 * x1ii - 0.9

# iii values
x1iii = np.linspace(setosa_df['sepal_length (cm)'].min(), setosa_df['sepal_length (cm)'].max(), 100)
x2iii = -0.2 * x1iii + 12

# mapping each class label to a color
colors = setosa_df['label'].map(lambda x: 'orange' if x == 'setosa' else 'blue')

# creating a scatter plot
plt.scatter(setosa_df['sepal_length (cm)'], setosa_df['sepal_width (cm)'], c=colors)

# plot decision boundary 1 2 and 3
plt.plot(x1i, x2i, color='skyblue', label="Decision Boundary 1")
plt.plot(x1ii, x2ii, color='blue', label="Decision Boundary 2")
plt.plot(x1iii, x2iii, color='red', label="Decision Boundary 3")

# plotting details
plt.title("Decision Boundary 1, 2 and 3 Plots")
plt.xlabel("sepal_length (cm)")
plt.ylabel("sepal_width (cm)")
plt.legend()
plt.show()  
```

```python

"""

    e. rewrite each decision boundary equation in a modified standard form
        w1x1 + w2x2 + w0 = 0
    
    f. import the linear_decision_boundary_classifier() function from the provided decision_boundaries.py file. Call the function for each
        decision boundary like so:
        
        dec_bound_vec = np.array([0.75, -1.0, -0.9])
        features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values
        true_labels = setosa_df["label"].values
        pred_labels = linear_decision_boundary_classifier(dec_bound_vec, features, true_labels, features)
        print(pred_labels)

"""

import sys
sys.path.append('./code')

# rewriting each decision boundary equation
    # i. x2 = 2 x1 - 8 --> 2 x1 - x2 - 8 = 0
    # ii. x2 = 0.75 x1 - 0.9 --> 0.75 x1 - x2 - 0.9 = 0
    # iii. x2 = -0.2 x1 + 12 --> -0.2 x1 - x2 + 12 = 0


# importing the linear decision boundary classifier
from decision_boundaries import linear_decision_boundary_classifier  # type: ignore


setosa_map = {'setosa' : 1, 'not setosa' : 0}
true_labels = setosa_df['label'].map(setosa_map).values
features = setosa_df[["sepal_length (cm)", "sepal_width (cm)"]].values

# decision boundary 1
dec_bound_vec_1 = np.array([2, -1, -8])
pred_labels_1 = linear_decision_boundary_classifier(dec_bound_vec_1, features, true_labels, features)

# print("Predicted Labels 1:")
# print(pred_labels_1, "\n")

# decision boundary 2
dec_bound_vec_2 = np.array([0.75, -1, -0.9])
pred_labels_2 = linear_decision_boundary_classifier(dec_bound_vec_2, features, true_labels, features)

# print("Predicted Labels 2")
# print(pred_labels_2, "\n")

# decision boundary 3
dec_bound_vec_3 = np.array([-0.2, -1, 12])
pred_labels_3 = linear_decision_boundary_classifier(dec_bound_vec_3, features, true_labels, features)

# print("Predicted Labels 3")
# print(pred_labels_3, "\n")
```

```python

"""

    g. use the scikit-learn sklearn.metrics.accuracy_score() function to calculate and print the accuracy of the predictions using each decision
        boundary using the true and predicted labels

"""

# importing required libraries
from sklearn.metrics import accuracy_score

true_labels = np.array(true_labels)

# numpy data type to int data type
pred_labels_1 = np.array(pred_labels_1, dtype=int)
pred_labels_2 = np.array(pred_labels_2, dtype=int)
pred_labels_3 = np.array(pred_labels_3, dtype=int)

# classification accuracy
accuracy_1 = accuracy_score(true_labels, pred_labels_1)
accuracy_2 = accuracy_score(true_labels, pred_labels_2)
accuracy_3 = accuracy_score(true_labels, pred_labels_3)

# printing out results
print("Accuracy of Decision Boundary 1:", f"{accuracy_1:.3f}")
print("Accuracy of Decision Boundary 2:", f"{accuracy_2:.3f}")
print("Accuracy of Decision Boundary 3:", f"{accuracy_3:.3f}")
```
