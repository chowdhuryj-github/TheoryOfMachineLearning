### The Gaussian Distribution Probability Density Function

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}$$


In this probability density function, the model parameters are $\mu$ and $\sigma$ The independent variables are x and the dependent variables are f(x).

### **Part 3:** Model Implementation
Implement the gaussian distribution model. The code should operate on the independent variable and "fixed" model parameters to produce the dependent variable as output. To start off, the code should be able to take a single independent variable (scalar) and produce a single dependent variable (scalar), but ultimately, it should support the use of vector input / output.

```python
# defining a method based on the gaussian distribution pdf
# ! need to change this to accept vectors

def gaussian_pdf_plot(input_val: np.ndarray, mu: float, sigma: float) -> np.ndarray:

    """
        this is a method that returns a numpy array of the predictions being made
    """

    part_one = 1 / (sigma * np.sqrt(2 * np.pi))
    part_two = -1 * (input_val - mu)**2 / (2 * sigma**2)
    final_val = part_one * (np.exp(part_two))
    
    return final_val
```

### **Part 4:** Model Output and Visualization
Use the coded model to generate model predictions (output) using the dataset's independent variable values. On the same figure, plot the model predictions from the following set of model parameters and the given data. 

```python
array_one = gaussian_pdf_plot( gauss_df["1"], 1, 0.75)
array_two = gaussian_pdf_plot( gauss_df["1"], 1, 1.25)
array_three = gaussian_pdf_plot( gauss_df["1"], 5, 1.25)
array_four = gaussian_pdf_plot( gauss_df["1"], 6, 1.25)

plt.scatter(gauss_df["1"], array_one, label="mu=1, sigma=0.75")
plt.scatter(gauss_df["1"], array_two, label="mu=1, sigma=1.25")
plt.scatter(gauss_df["1"], array_three, label="mu=5, sigma=1.25")
plt.scatter(gauss_df["1"], array_four, label="mu=6, sigma=1.25")

plt.title("The Gausssian PDF using Different Mu & Sigma Values")
plt.xlabel("Feature")
plt.ylabel("Predictions")

plt.legend()
plt.show()
```

### Part 5: Error Between Model and Data
Each plot that was produced depics the data that was given and data that the model produced for a given set of parameters. The goal is choose a set of parameters that minimizes the difference between the model output and the given data. The difference between those two sets is known as the model error.

A common metric used to quantify error is known as the Mean Absolute Error. We use this metric to characterize the error between the model and the given data. Here are the following deliverables:
- [x] Use markdown to write out the mathematical form of Mean Abbsolute Error. We use this metric to characterize the error between the model and the given data.
- [ ] For each set of model predictions (4 sets  1 for each parameter set), calculate the mean absolute error between the model prediction and the given data.
- [ ] Plot all of the model predictions and the given data in a new figure and add the calculated error values to the figure legend.

### The Mean Absolute Error Formula

$$ MAE = \frac{1}{n} \sum_{i=1}^{n} | x_i - x   |  $$

```python
# defining a function that calculates the mean absolte error

def compute_mae(array_one: np.ndarray, array_two: np.ndarray) -> float:

    """
        this is a method that computes the mean absolute error and returns it as a float
    """

    difference = np.abs(np.subtract(array_one, array_two))
    difference_sum = np.sum(difference)
    mean_square_error = (1 / len(array_one)) * difference_sum
    return mean_square_error
```

```python
# the mean absolute error list
mae_list = []

# computing the mean absolute error and appending
mae_list.append(compute_mae(gauss_df["2"], array_one))
mae_list.append(compute_mae(gauss_df["2"], array_two))
mae_list.append(compute_mae(gauss_df["2"], array_three))
mae_list.append(compute_mae(gauss_df["2"], array_four))

# printing out the results
print("All Mean Absolute Errors: ", mae_list)
```

```python
# plot 1: model predictions, given data and mae

plt.scatter(gauss_df["1"], gauss_df["2"], label="original")
plt.scatter(gauss_df["1"], array_one, label=f"MAE={mae_list[0]:.3f}")
plt.scatter(gauss_df["1"], array_two, label=f"MAE={mae_list[1]:.3f}")
plt.scatter(gauss_df["1"], array_three, label=f"MAE={mae_list[2]:.3f}")
plt.scatter(gauss_df["1"], array_four, label=f"MAE={mae_list[3]:.3f}")


plt.title("The Gaussian PDF using Different Mu & Sigma Values")
plt.legend()
plt.show()
```
